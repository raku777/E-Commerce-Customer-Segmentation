```python
new_df = df.copy()
new_df.head(10)
```

```python
new_df.iloc[:,3:].sum(axis=1)
```

The above values are the sum of total searched done by each customers ID. This can be used as a feature

```python
new_df['Total_Search'] = new_df.iloc[:,3:].sum(axis=1)
new_df.head()
```

```python
plt.figure(figsize=(10,5))
#
plt_data = new_df.sort_values('Total_Search')[['Cust_ID','Gender','Total_Search']]
#
sns.scatterplot(plt_data, x = 'Total_Search', y = 'Cust_ID',hue = 'Gender')
```

Lets call the array which we are gonna train as x

```python
x = new_df.iloc[:,2:].values
x
```

```python
x.shape
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score
import yellowbrick
from yellowbrick.cluster import KElbowVisualizer
```

```python
features = MinMaxScaler().fit_transform(x)
features
```

```python
inertia = []
for i in range(1,16): 
    k_means = KMeans(n_clusters=i)
    
    k_means = k_means.fit(features)
    
    inertia.append(k_means.inertia_)
```

```python
plt.figure(figsize=(20,7))

plt.subplot(1,2,1)
plt.plot(range(1,16), inertia, 'bo-')

plt.xlabel('No of clusters'), plt.ylabel('Inertia')

plt.subplot(1,2,2)

kmeans = KMeans()

visualize = KElbowVisualizer(kmeans,k=(1,16)) 
visualize.fit(features)

plt.suptitle("Elbow Graph and Elbow Visualizer")
visualize.poof()

plt.show()
```

K=3 seems to be the optimum number of clusters 


```python
model = KMeans(n_clusters=3)
model = model.fit(features)
```

```python
X = model.predict(features)
centers = model.cluster_centers_
centers
```

```python
X
```

```python
new_df['Cluster']=pd.DataFrame(X)
new_df.head()
```

```python
new_df['Cluster'].value_counts()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data=new_df,x='Cluster')
plt.show()
```

Cluster 0

```python
cl_0 = new_df.groupby(['Cluster','Gender'],as_index=False).sum().query('Cluster == 0')
cl_0
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
sns.countplot(data=new_df.query('Cluster == 0'),x='Gender')
plt.title('Customers count')

plt.subplot(1,3,2)
sns.barplot(data=cl_0,x='Gender',y='Total_Search')
plt.title('Total Searches by Gender')
plt.suptitle('No. of Customers and their Total Searches in "Cluster 0"')

plt.subplot(1,3,3)
sns.barplot(data=cl_0,x='Orders',y='Total_Search',hue='Gender')
plt.title('Order Count')
plt.show()
```
